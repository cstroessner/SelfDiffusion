In this notebook, is coded a stochastic method in order to compute the mean square deviation of the tagged particle process in 2D. 

```julia
import Pkg; Pkg.add("LaTeXStrings")
using ForwardDiff
using LaTeXStrings
using NLsolve
using Plots
using Random
```

Definition of the lattice, jumps and jump rates

The lattice is {-N, ..., 0, ... N}^2

```julia
N = 2

jump_list = [ [1,0], [0,1], [-1,0], [0, -1]]
jump_rates = [0.25, 0.25, 0.25, 0.25]
K = 4 ## Number of jump directions

sum_rates = sum(jump_rates)
jump_proba = jump_rates/sum_rates

# Give the number of the case whose lower left corner is located at (i/N, j/N)
function ind_case(i,j)
    return (i+N+1) + (2N+1)*(j+N)
end

# Gives the coordinate of the lower left corner of the k^th case
function coord_case(k)
    j = Int(floor((k-1)/(2N+1)) - N)
    i = k - (2N+1)*(j+N) - N -1
    return i,j
end

#Gives the (periodized) index of a point in the grid()
function periodize(i,j)
    inew = i
    jnew = j
    if (i >N)
        inew = i - (2*N+1)
    end
    if (i < -N)
        inew = i + (2*N+1)
    end
    if (j >N)
        jnew = j - (2*N+1)
    end
    if (j < -N)
        jnew = j + (2*N+1)
    end
    return inew, jnew
end

function show_occupied(list)
    X = []
    Y = []
   for ind= 1:length(list)
        i,j = coord_case(list[ind])
        append!(X,i)
        append!(Y,j)
    end
    return X, Y 
end
```

Initial state of the environment

```julia
l = 3 #l is the number of occupied sites (excluding the site of the tagged particle, which is always occupied)

occupancy_tab = zeros(Int, (2N+1)*(2N+1))
occupancy_tab[ind_case(0,0)] =1; 

#Pick l random sites except the tagged particle one

Sites_list = zeros(Int, (2N+1)*(2N+1))
for i = 1:((2N+1)*(2N+1))
    Sites_list[i] = i
end

deleteat!(Sites_list, ind_case(0,0))
len = (2N+1)*(2N+1) -1

occ_sites = [ind_case(0,0)]

for L = 1:l
    
    s = rand();
    k = Int(ceil(len*s))
    occupancy_tab[Sites_list[k]] =1;
    
    append!(occ_sites, Sites_list[k])
    deleteat!(Sites_list, k)
    
    len = len-1
end

println(occ_sites)
println(occupancy_tab)

X,Y = show_occupied(occ_sites)
scatter(X,Y)
```

Draw times for jumping events: they all follow an exponential law with rate sum_rates

```julia
time_list = [0.0]
len = length(time_list)

final_time  = 10000.0; #Parameter to change to see the evolution

while (time_list[len] < final_time)
    dtnew = (1.0/sum_rates)*randexp(); 
    tnew = time_list[len] + dtnew; 
    append!(time_list, tnew); 
    len = len+1
    
end

print(time_list)
```
